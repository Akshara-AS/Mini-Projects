```python
plt.figure(figsize=(15, 12))  # Set the figure size
sns.heatmap(df.corr(), annot=True, fmt=".2f", linewidths=0.5,vmax=0.5,annot_kws={'fontsize':7})
plt.title('Correlation Heatmap', fontsize=14)
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels for better visibility
plt.yticks(rotation=0)
plt.show()
```

```python
correlation_matrix = df.corr()

# Extract the correlations of each column with the target variable
correlations = correlation_matrix['h1n1_vaccine'].abs().sort_values(ascending=False).round(2)

# Display the correlations in decreasing order
print(correlations)
```

dependent_variable = 'h1n1_vaccine'  # Replace 'target' with the name of your dependent variable column
numerical_columns = df.select_dtypes(include='number').columns

#Calculate correlation with numerical columns
numerical_correlations = df[numerical_columns].corr()[dependent_variable].sort_values(ascending=False).round(2)

#Display the correlation coefficients
print(numerical_correlations)

```python
categorical_columns = df.select_dtypes(include='object').columns

# Calculate correlation with categorical columns using chi-square test
categorical_correlations = {}

for column in categorical_columns:
    contingency_table = pd.crosstab(df[column], df['h1n1_vaccine'])
    _, p_value, _, _ = chi2_contingency(contingency_table)
    categorical_correlations[column] = round(p_value,2)

# Sort the categorical correlations by descending p-values
sorted_correlations = pd.Series(categorical_correlations).sort_values(ascending=False)

#sorted_correlations = sorted(categorical_correlations.items(), key=lambda x: x[1], reverse=True)

# Display the sorted correlations
#for column, p_value in sorted_correlations:
#    print(f"{column}: {p_value}")

print(sorted_correlations)
```

```python
correlations = correlations.append(sorted_correlations).sort_values(ascending=False)
correlations
```

```python
correlations.index[correlations.values < 0.1]
```

```python
threshold = 0.10

# Find the columns to drop
columns_to_drop = correlations.index[correlations.values < threshold]

# Drop the columns from the DataFrame
df_dropped = df.drop(columns_to_drop, axis=1)

# Display the resulting DataFrame
df_dropped.shape #19 columns removed
```

```python
cols = ['h1n1_worry', 'h1n1_awareness', 'antiviral_medication',
       'contact_avoidance', 'bought_face_mask', 'wash_hands_frequently',
       'avoid_large_gatherings', 'reduced_outside_home_cont',
       'avoid_touch_face', 'dr_recc_h1n1_vacc', 'dr_recc_seasonal_vacc',
       'chronic_medic_condition', 'cont_child_undr_6_mnths',
       'is_health_worker', 'has_health_insur', 'is_h1n1_vacc_effective',
       'is_h1n1_risky', 'sick_from_h1n1_vacc', 'is_seas_vacc_effective',
       'is_seas_risky', 'sick_from_seas_vacc','no_of_adults', 'no_of_children']

for i in cols:
    df[i] = df[i].astype(str)  
    
```

```python
df['age_bracket'] = df['age_bracket'].replace({'18 - 34 Years' :'0','35 - 44 Years':'1','45 - 54 Years':'2','55 - 64 Years':'3','65+ Years':'4'})
df['qualification'] = df['qualification'].replace({'< 12 Years' :'0','12 Years':'1','Some College':'2','College Graduate':'3'})
df['race'] = df['race'].replace({'White' :'0','Black':'1','Other or Multiple':'2','Hispanic':'3'})
df['sex'] = df['sex'].replace({'Female' :'0','Male':'1'})
df['income_level'] = df['income_level'].replace({'Below Poverty' :'0','<= $75,000, Above Poverty':'1','> $75,000':'2'})
df['marital_status'] = df['marital_status'].replace({'Not Married' :'0','Married':'1'})
df['housing_status'] = df['housing_status'].replace({'Own' :'0','Rent':'1'})
df['employment'] = df['employment'].replace({'Not in Labor Force' :'0','Unemployed':'1','Employed':'2'})
df['census_msa'] = df['census_msa'].replace({'Non-MSA' :'0','MSA, Not Principle  City':'1','MSA, Principle City':'2'})
```

## Model Building

```python
X = df.drop('h1n1_vaccine',axis=1)
y = df['h1n1_vaccine']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=0)
```

## Logistic Regression

```python
model_1 = LogisticRegression()
model_1.fit(X_train,y_train)
```

```python
model_1.score(X_train,y_train)
```

```python
model_1.score(X_test,y_test)
```

```python
predictions = model_1.predict(X_test)
```

```python
cm = metrics.confusion_matrix(y_test,predictions,labels=[1,0])
df_cm = pd.DataFrame(cm,index=[i for i in ['1','0']], columns = [i for i in ['Predict 1','Predict 0']])
plt.figure(figsize=(7,5))
sns.heatmap(df_cm,annot=True,fmt='g')
```

```python
accuracy_score(y_test,predictions)
```

```python
print(metrics.classification_report(y_test,predictions))
```

## KNN Classifier

```python
model_2 = KNeighborsClassifier(n_neighbors=15)
model_2.fit(X_train,y_train)
```

```python
model_2.score(X_train,y_train)
```

```python
model_2.score(X_test,y_test)
```

```python
predictions = model_2.predict(X_test)
```

```python
cm = metrics.confusion_matrix(y_test,predictions,labels=[1,0])
df_cm = pd.DataFrame(cm,index=[i for i in ['1','0']], columns = [i for i in ['Predict 1','Predict 0']])
plt.figure(figsize=(7,5))
sns.heatmap(df_cm,annot=True,fmt='g')
```

```python
accuracy_score(y_test,predictions)
```

```python
print(metrics.classification_report(y_test,predictions))
```
